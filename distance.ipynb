```python
import dhlab as dh
import pandas as pd
import requests
from collections import Counter
```

```python
def get_text(urn):
    t =  requests.get("http://sprakbankdb2.lx.nb.no:7007/text", params={'identifier':urn})
    if t.status_code == 200:
        res = t.text
    else:
        res = 'NaT'
    return res.split()
```
