```python
import dhlab as dh
import pandas as pd
import requests
from collections import Counter
```

```python
def get_text(urn):
    t =  requests.get("http://sprakbankdb2.lx.nb.no:7007/text", params={'identifier':urn})
    if t.status_code == 200:
        res = t.text
    else:
        res = 'NaT'
    return res.split()
```

```python
c = dh.Corpus(doctype="digibok", from_year = 1980, limit = 100)
```

```python
c.frame.urn    [0]
```

```python
a = get_text(c.frame.urn[0])
```

```python
indexed = [(i,e) for i,e in enumerate(a)]
```

```python
ip = {e: pd.Series([i for (i,el) in indexed if e == el]).describe() for e in set(a)}
```

```python
def check_book(a, ip, devs, stds, counts):
    bits = len(a)/100
    mid = len(a)/2
    words = [e for e in ip if mid-devs*bits <= ip[e]['mean'] <= mid+devs*bits and ip[e]['std'] <= stds*bits and ip[e]['count'] >=counts ]
    
    return pd.concat([pd.DataFrame(ip[x], columns=[x]) for x in words], axis = 1).transpose()

def check_std(a, ip, stds, counts):
    bits = len(a)/100
    mid = len(a)/2
    words = [e for e in ip if  ip[e]['std'] <= stds*bits and ip[e]['count'] >=counts ]
    
    return pd.concat([pd.DataFrame(ip[x], columns=[x]) for x in words], axis = 1).transpose()
```

```python
check_std(a, ip, 30, 5 ).sort_values(by='std').style
```
